# Merging two dataset

```python
final_df['Country'].value_counts()
```

```python
country_name = final_df['Country'].value_counts().index
country_name 
```

```python
country_value = final_df['Country'].value_counts().values
country_value
```

# Finding Top 5 countries with highest transaction


```python
df= pd.DataFrame(country_value, country_name)
df.rename(columns = {0: 'value'})[0:5]
```

```python

plt.pie(country_value[:3], labels= country_name[:3], autopct = "'%0.1f%%'")
```

## Grouping and finding relation between variables

```python


ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings

```

```python
matplotlib.rcParams['figure.figsize'] =(10,6)
sns.barplot(x='Aggregate rating', y = 'Rating Count', data =ratings)
```

```python
bins = np.linspace(0,5, 6)
group_name = ['not rated','poor','average','good','excellent']
ratings['ratings'] = pd.cut(ratings['Aggregate rating'], bins, labels = group_name, include_lowest = True)
ratings
```

```python
matplotlib.rcParams['figure.figsize'] =(10,4)
sns.countplot(x='ratings', data =ratings)
```
